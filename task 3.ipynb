# Water Jug Problem using DFS

```python
def dfs(jug1, jug2, target, visited, path):
    if (jug1, jug2) in visited:
        return False
    
    visited.add((jug1, jug2))
    path.append((jug1, jug2))
    
    print(f"Current State: {jug1}L , {jug2}L")
    
    if jug1 == target or jug2 == target:
        print("Target reached!")
        return True
    
    # Possible operations
    operations = [
        (4, jug2),        # Fill Jug1
        (jug1, 3),        # Fill Jug2
        (0, jug2),        # Empty Jug1
        (jug1, 0),        # Empty Jug2
        (jug1 - min(jug1, 3-jug2), jug2 + min(jug1, 3-jug2)),  # Pour Jug1 → Jug2
        (jug1 + min(jug2, 4-jug1), jug2 - min(jug2, 4-jug1))   # Pour Jug2 → Jug1
    ]
    
    for state in operations:
        if dfs(state[0], state[1], target, visited, path):
            return True
    
    path.pop()
    return False

visited = set()
dfs(0, 0, 2, visited, [])
```
